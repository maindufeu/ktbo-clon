```r
#Modificar la fecha de incio
df[,'Start Date (DD/MM/YYYY)':= as.Date(format(df[,'Start Date (DD/MM/YYYY)'], "%Y/%m/01"))]
```

```r
#Crea la función month_duration para modificar la fecha. Es necesario tener en cuenta que el argumento debe ser un string y el output será un string

month_durationdate = function(date) {
    date1 = as.Date(date)
    end_day = 0
    month = as.integer(format(date1, "%m"))
    if ((month) %in% c(1,3,5,7,8,10,12)) end_day = 31
    else if ((month) %in% c(4,6,9,11)) end_day = 30
    else if ((month) == 2) end_day = 29
    str_date = paste((format(date1, "%Y-%m")),end_day, sep = "-")
    return(str_date)
}
```

```r
#Cambia el formato
df$End.Date = format(df$'End Date (DD/MM/YYYY)', "%Y-%m-%d")
```

```r
#Aplica la función y la muestra
df$End.Date = as.Date(unlist(lapply(df$End.Date, month_durationdate)))
df[, c("Start Date (DD/MM/YYYY)", "End Date (DD/MM/YYYY)", "End.Date")]
```

```r
df$'End Date (DD/MM/YYYY)'=as.Date(df$End.Date)
df[,c('End Date (DD/MM/YYYY)', 'End.Date')]
```

```r
write.xlsx(df, "MCMA_Mediaplan_Acumulado_20200101to20200831.xlsx", sheetName = "MEDIA PLAN", startRow = 11)
```
